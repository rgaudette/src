```python
from matplotlib.pyplot import *
from numpy import *

from matplotlib import rcParams
print(rcParams["figure.figsize"])
print(rcParams["figure.dpi"])
print(rcParams["font.size"])
print(rcParams["lines.linewidth"])
# print(rcParams.keys())
rcParams["figure.figsize"] = (8, 4)
# Accurate for howlin under QT
rcParams["figure.dpi"] = 187
rcParams["figure.facecolor"] = "gray"
rcParams["figure.edgecolor"] = "gray"
rcParams["lines.linewidth"] = 1.0
rcParams["lines.markersize"] = 3.0
rcParams["grid.linewidth"] = 0.5
rcParams["font.size"] = 8
v = arange(8)
v_sq = v ** 2
fig, ax = subplots(1, 1)
plot(v, "bo-", label="linear")
plot(v_sq, "ro-", label="squared")
ax.text(0, 30, "axes text", horizontalalignment="left", verticalalignment="bottom")
grid(True)
xlabel("X Label")
ylabel("Y Label")
title("Title")
legend();
fig.tight_layout()
fig.text(0, 0, "figure text", horizontalalignment="left", verticalalignment="bottom");
```
