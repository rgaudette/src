# Jupyter Julia Notebooks PyPlot Graphics Properties
This document identifies the propteries of Jupyter Julia notebooks using the PyPlot module

```julia
using PyCall
using PyPlot
pygui(:qt5)
# using PyPlot
println("The default figure DPI is ", matplotlib.rcParams["figure.dpi"])
```

For a 5" x 5" figure, this generates an 8.75" figure

```julia
real_dpi = 187 / (8.75)  * 5
println(matplotlib.get_backend())
#'TKAgg','GTKAgg','Qt4Agg','WXAgg'
# matplotlib.pyplot.switch_backend("Qt4Agg")
# println(matplotlib.get_backend())
```

```julia
PyCall.PyDict(matplotlib["rcParams"])["figure.dpi"] = [real_dpi]
f = figure(figsize=(5, 5))
plt.plot(range(1, 10, length=10))
plt.grid(true)
# plt.show()
```
