```julia
# offline_data_path_name = raw"C:\AXI\Offline Data"

# dataset_name = raw"Walker Pass\WKPs1_M11"
# log_sub_directory = raw"irpLog\IRP@9119_2021_Nov_30_1722_31\data"

# Abraham Creek\Abraham_il2\Abraham_il2_speed1\irpLog_canonlake_il2_speed1\IRP@9119_2020_Feb_03_1501_32\data\cam_00_00_0_0_1087_862.png
# Backdrill\2021-03-30\irpLog\IRP@9121_2021_Mar_30_0903_52\data\cam_00_00_0_0_1087_905.png
# Intel Pear Lake\XLT\PearLake_M15\irpLog_PearLake_M15\IRP@9120_2021_Apr_14_1053_54\data\cam_00_00_0_0_1087_676.png
# Intel Pear Lake\Intel Peal Lake DRO3 2018-02-26\dro3\1\data\cam_00_00_0_0_1087_3080.png
# continental\2018-03-16\1\data\cam_00_00_0_0_1087_754.png
# VIE Board\VIE_board_E1\1\data\cam_00_00_0_0_1360_4037.png

# ViTrox Demo Board\2017-08-06\projections\cam_00_00_0_3089_1087_5159.png
# Steel ball jig\steelball_m19_il2_speed1\projections\cam_00_00_0_0_1087_534.png

function find_irp_directories(path, irp_sets=nothing)
    # First time through, create the irp_sets dictionary
    if irp_sets === nothing
        irp_sets = Dict{String, Vector{String}}()
    end
    n_root_tokens = length(split(path, "\\"))
    # walkdir will iterative walk the whole tree under path
    for (pathname, dirs, files) in walkdir(path)
        println("pathname: ", pathname)
        # Search the list of files for one that begins with "cam_00_00", this indicates the currect directory is an IRP
        # projections directory. Parse the pathname to extract the dataset_name and log_sub_directory
        for file in files
            # println("file: ", file)
            if startswith(file, "cam_00_00")
                tokens = split(pathname, "\\")
                if tokens[end] == "projections"
                    if length(tokens) < 4
                        throw(ErrorException("Insuffucient number of sub-directories in $pathname"))
                    end
                    dataset_name = joinpath(dataset_name[n_root_tokens+1:end-1])
                    log_sub_directory = tokens[end]
                else
                    if length(tokens) < 5
                        throw(ErrorException("Insuffucient number of sub-directories in $pathname"))
                    end
                    dataset_name = joinpath(tokens[n_root_tokens+1:end-2])
                    log_sub_directory = joinpath(tokens[end-1:end])
                end
                println("  dataset_name: ", dataset_name)
                println("  log_sub_directory: ", log_sub_directory)
                if haskey(irp_sets, dataset_name)
                    push!(irp_sets[dataset_name], log_sub_directory)
                else
                    irp_sets[dataset_name] = [log_sub_directory]
                end
            end
        end
    end

    return irp_sets
end

root = raw"C:\Users\rick\temp"
irp_sets = find_irp_directories(root)
for (ds_name, irp_dirs) in irp_sets
    println(ds_name)
    for irp_dir in irp_dirs
        println("  ", irp_dir)
        test_path = joinpath(root, ds_name, irp_dir)
        print(test_path)
        if ispath(test_path)
            println(" exists")
        else
            println(" DOES NOT EXIST")
        end
    end
end
        
```
